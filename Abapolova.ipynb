```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('tele.csv',  delimiter = ',')
data
```

```python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'],errors='coerce')
data.info() 

# Пропусков в данных  нет
```

```python
columns_of_many_val  = []
for _ in range(data.shape[1]):
    name_col =data.columns[_]
    count  = len(list(data[name_col].value_counts()))
    if count > 5: #множество различных  значений этого признака
        print(name_col)
        columns_of_many_val.append(name_col)
    
```

```python
for i in  columns_of_many_val[1::]:
    q = data[i].quantile(0.99)
    data = data[data[i] < q]
data
```

```python


for _ in range(data.shape[1]):
    name_col =data.columns[_]
    if data[name_col].dtype == 'O':
        data[name_col] =data[name_col].astype('category')

data.info()
        
```

```python
corr = data.apply(lambda x : pd.factorize(x)[0]).corr(method='pearson', min_periods=1)

corr
```

```python
сhurn = corr['Churn'].to_dict()
сhurn
```

```python
sorted_values = sorted(сhurn.values()) # Sort the values
sorted_dict = {}
sorted_values  = [abs(el) for el in sorted_values]
sorted_values = sorted(sorted_values)
for i in sorted_values:
    for k in сhurn.keys():
        if round(abs(сhurn[k]), ndigits=5) == round(abs(i), ndigits=5):
            sorted_dict[k] = сhurn[k]
            break

print(sorted_dict)
```
