```python
import pandas as pd
import numpy as np
import decimal
```

```python
data = pd.read_csv('empl.csv', delimiter = ',')
data
```

```python
data.info()
```

```python
data = pd.get_dummies(data,columns=['BusinessTravel','Department','EducationField'])
data
```

```python
data.replace({'Attrition':{'Yes':1, 'No':0}}, inplace =True)
data
```

```python
corr = data.corr()
corr
```

```python
data.drop(['StandardHours','EmployeeCount'], axis=1, inplace =True)
corr = data.corr()
job = corr['JobSatisfaction'].to_dict()
job
```

```python
sorted_values = sorted(job.values()) # Sort the values
sorted_dict = {}
sorted_values  = [abs(el) for el in sorted_values]
sorted_values = sorted(sorted_values)
for i in sorted_values:
    for k in job.keys():
        if round(abs(job[k]), ndigits=5) == round(abs(i), ndigits=5):
            sorted_dict[k] = job[k]
            break

print(sorted_dict)
```
